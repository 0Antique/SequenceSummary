```python
import json
import glob
import os

# 读取所有训练集 JSON 文件
train_files = glob.glob('data/train/train_*.json')
print(f"找到 {len(train_files)} 个训练文件")

# 统计符合条件的记录: domain=Travel, subdomain=Airlines, website=delta
delta_count = 0
total_count = 0

for file_path in train_files:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
        for record in data:
            total_count += 1
            # 检查是否满足: domain=Travel, subdomain=Airlines, website=delta
            domain = record.get('domain', '').lower()
            subdomain = record.get('subdomain', '').lower()
            website = record.get('website', '').lower()
            
            if 'travel' in domain and 'airlines' in subdomain and 'delta' in website:
                delta_count += 1

print(f"\n总记录数: {total_count}")
print(f"符合条件的记录数 (domain=Travel, subdomain=Airlines, website=delta): {delta_count}")
print(f"占比: {delta_count/total_count*100:.2f}%")
```

```python
# save as tools/reencode_json_utf8.py and run with python
import pathlib

assets = pathlib.Path("visualization/app/public/assets")
for p in assets.glob("*.json"):
    try:
        # 常见中文环境原文件编码为 gbk/gb2312
        text = p.read_text(encoding="gbk")
    except UnicodeDecodeError:
        # 若不是 gbk，尝试按系统默认或 utf-8 读取
        try:
            text = p.read_text(encoding="utf-8")
        except UnicodeDecodeError:
            text = p.read_text(encoding="mbcs", errors="replace")
    p.write_text(text, encoding="utf-8")
    print("converted ->", p)
```
